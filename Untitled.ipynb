### T-test of prices of houses with/without waterfront

```python
# split the dataframe (1 split for houses with a waterfront and the other for houses without a waterfront)
df_water0= df_lr[df_lr['waterfront']== 0]['price']
df_water1= df_lr[df_lr['waterfront']== 1]['price']
```

H0 = 'mu1 = mu0' <br>
Ha= 'mu1 > mu0' <br>
alpha= 0.05

```python
# checking if variance is equal
print(df_water0.std())
print(df_water1.std())
```

```python
stats.ttest_ind(df_water1, df_water0, equal_var= False, alternative= 'greater')
```

### Results of t-test
With the p-value so much lower than the confidence level, we reject the null hypothesis, which means that it is highly unlikely that the presence of a waterfront at a home does't greatly increases the home's value.

## Modeling

### Base model

```python
fig, ax = plt.subplots(figsize=(20, 10))
sns.scatterplot(x= df_lr['sqft_living'], y= df_lr['price'])
plt.suptitle("Price against sqft_living")
sns.set(font_scale = 3)
```

```python
# checking if dist sqft_living is normal
fig, ax = plt.subplots(figsize=(10, 5))
sns.set(font_scale = 1)
sns.distplot(df_lr['sqft_living']);
```

The distribution is normal, therefore log transformation is unnecessary

```python
# train_test_split the dataframe
y= df_lr['price']
x= df_lr.drop(['price','id','date'], axis= 1)
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
train= pd.concat([x_train, y_train], axis= 1)
test= pd.concat([x_test, y_test], axis= 1)
```

```python
formula= "price ~ sqft_living"
model = ols(formula, data= train).fit()
model.summary()
```

The R-squared value is 0.499 which isn't very good. As you can see in the plot below, the model isnt very accurate. This inaccuracy is probably present because the plot displays <b>Heteroscedasticity</b>.

```python
fig, ax= plt.subplots(figsize= (20,10))
sns.set(font_scale = 3)
sm.graphics.plot_fit(model,1, vlines=False, ax= ax);
```

### Perform log transformations

```python
# checking skewness to find non-normal distributions
to_transform= []
for i in x_train.columns:
    print(i+ ": "+str(x_train[i].skew()))
    if x_train[i].skew() > 1 or x_train[i].skew() <-1:
        to_transform.append(i)
```

```python
# removing log values that go to infinity
for i in ['waterfront', 'view', 'sqft_basement', 'yr_renovated']:
    to_transform.remove(i)
to_transform
```

```python
for i in to_transform:
    train[i] = train[i].map(lambda x: np.log(x))
```

```python
formula= """price ~ bedrooms + bathrooms + sqft_living + sqft_lot + C(waterfront) + view + condition+
            grade + sqft_above + yr_built + yr_renovated + zipcode + sqft_living15 + sqft_lot15"""
fourth_model = ols(formula, data= train).fit()
fourth_model.summary()
```
